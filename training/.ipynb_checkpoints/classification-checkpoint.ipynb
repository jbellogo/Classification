```python
### Prepare Ed data
def prepare_df(df, suffix : str):
    uids = df['psu'].astype(str) + "_"  + df['hh'].astype(str) + "_" + df['idcode'].astype(str) 
    df =  df.drop(columns=['psu', 'hh', 'idcode'])
    df.columns = [suffix + '_' + col for col in df.columns]
    df.insert(0, 'uid', uids)
    return df
    
# uids = ed_df['psu'].astype(str) + "_"  + ed_df['hh'].astype(str) + "_" + ed_df['idcode'].astype(str) 
# ed_df.insert(0, 'uid', uids)
# ed_df =  ed_df.drop(columns=['psu', 'hh', 'idcode'])
# ed_df.head()

ed_df = prepare_df(ed_df, "Ed")
hh_df = prepare_df(hh_df, "Hh")
```

### Which columns can be replaced by a zero?

* Categorical answered-questions for which zero is not one of the categories. By replacing them with a zero, we are implicitly adding an additional category for "No response"
  
* We need to make sure zero is not a feasible value in the column. otherwise it already means something and we need a new indicator. We also need to make sure it is categorical, if a column is numeric and we add a value of zero, then that doesnt make sense, add the average instead. 

```python
min_vals = df[columns_with_nan].min()
zero_valued_columns = []
non_zero_valued_columns = []

for index, value in min_vals.items():
    if float(value) < 1:
        zero_valued_columns.append(index)
        # print(f"Index: {index}, Value: {value}")
    else:
        non_zero_valued_columns.append(index)
```

# Cleaning up Education

```python
df = pd.read_csv('education.csv') # education
df = prepare_df(df, "Ed")
```

## Mergining Ed Q8, Q9, Q10, Q11 into one
about attending school
* Q8 Did you enroll this year?									
* Q9 will you attend this year?									
* Q10 Why not attending this year?							
* Q11 Why not enrolling this year?

Funny situations:
enrolled but will not attend. We can only care about attending. 

Our aggregation variable could be a merge of "why not attending/enrolling or attending"

After merging all 4 columns into one, we may need to further break down, transform the categories... maybe get rid of anything that is not about money (illness, poor quality of teaching, )

```python
######### Q10 and Q11
## the values match except for 13. An answer of 13 in Q10 should be transformed into a 14 before merging 
df.loc[df['Ed_Q10'] == 13, 'Ed_Q10'] = 14

# RANGES
# print(df['Ed_Q10'].describe()) # range 1-13
# print('\n')
# print(df['Ed_Q11'].describe())  # range 1-14

# filtered = df[df['Ed_Q10'] == 13]
# filtered['Ed_Q10']
```

```python
## any point when Ed_Q10 and Ed_Q11 are the not the same and are not NaN? NO
different_values = df['Ed_Q10'].ne(df['Ed_Q11']) & df['Ed_Q11'].notna() & df['Ed_Q10'].notna()
result_df = df.loc[different_values, ['uid','Ed_Q10', 'Ed_Q11']]
assert(len(result_df) == 0 )

# The fact that result_df has no entries means we can just merge them!!!!
df['Merged_Q10_Q11'] = df['Ed_Q10'].fillna(df['Ed_Q11'])
```

```python
########## Q8 and Q9
## This is the funny sitautation in which somene enrolled but will not attend. We only care about the fact that they will not attend

different_values = df['Ed_Q08'].ne(df['Ed_Q09']) & df['Ed_Q08'].notna() & df['Ed_Q09'].notna()
# Filter the DataFrame to get 'Ed_Q08' and 'Ed_Q09' only if they are different
result_df = df.loc[different_values, ['uid','Ed_Q08', 'Ed_Q09']]
result_df.head(3)

### can somne attend who did not enroll? NO... great!
different_values = df['Ed_Q08'].eq(2) & df['Ed_Q08'].ne(df['Ed_Q09']) & df['Ed_Q08'].notna() & df['Ed_Q09'].notna()
assert(len(df[different_values]) == 0)

```

```python
### Claim: if Merged_Q10_Q11 is NaN, then Ed_Q09 ==1. If they have no reason for not attending/enrolling, they will attend.
different_values = df['Merged_Q10_Q11'].isna() & df['Ed_Q09'].eq(2)
result_df = df.loc[different_values, ['uid','Merged_Q10_Q11', 'Ed_Q09']]
result_df.head(3)
assert(len(result_df)==0)
### YES

# Q9 dominates Q8 ie Merged_Q8_Q9 = Q9. So we merge Q9 with merged_Q10_Q11 by setting to zero NaN values in merged Q10_Q11
df['Ed_merged_Q8-11'] = df['Merged_Q10_Q11'].fillna(0)

df =  df.drop(columns=['Merged_Q10_Q11', 'Ed_Q08','Ed_Q09', 'Ed_Q10', 'Ed_Q11'])
```
