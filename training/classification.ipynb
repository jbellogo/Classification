```python
import pandas as pd
import os, sys

new_dir = '/Users/juanbello/Desktop/Kaggle'

sys.path.append(os.path.join(new_dir))

# Load the datasets
ed_df = pd.read_csv('education.csv') # education
hh_df = pd.read_csv('household.csv') # houshold

# # Display the first few rows of each dataset
ed_df.head()
# print(household_data.head())



```

```python
### Prepare Ed data
def prepare_df(df, suffix : str):
    uids = df['psu'].astype(str) + "_"  + df['hh'].astype(str) + "_" + df['idcode'].astype(str) 
    df =  df.drop(columns=['psu', 'hh', 'idcode'])
    df.columns = [suffix + '_' + col for col in df.columns]
    df.insert(0, 'uid', uids)
    return df
    
# uids = ed_df['psu'].astype(str) + "_"  + ed_df['hh'].astype(str) + "_" + ed_df['idcode'].astype(str) 
# ed_df.insert(0, 'uid', uids)
# ed_df =  ed_df.drop(columns=['psu', 'hh', 'idcode'])
# ed_df.head()

ed_df = prepare_df(ed_df, "Ed")
hh_df = prepare_df(hh_df, "Hh")
```

### Which columns can be replaced by a zero?

* Categorical answered-questions for which zero is not one of the categories. By replacing them with a zero, we are implicitly adding an additional category for "No response"
  
* We need to make sure zero is not a feasible value in the column. otherwise it already means something and we need a new indicator. We also need to make sure it is categorical, if a column is numeric and we add a value of zero, then that doesnt make sense, add the average instead. 

```python
min_vals = df[columns_with_nan].min()
zero_valued_columns = []
non_zero_valued_columns = []

for index, value in min_vals.items():
    if float(value) < 1:
        zero_valued_columns.append(index)
        # print(f"Index: {index}, Value: {value}")
    else:
        non_zero_valued_columns.append(index)
```

```python
######### Q10 and Q11
## the values match except for 13. An answer of 13 in Q10 should be transformed into a 14 before merging 
df.loc[df['Ed_Q10'] == 13, 'Ed_Q10'] = 14

# RANGES
print(df['Ed_Q10'].describe()) # range 1-13
print('\n')
print(df['Ed_Q11'].describe())  # range 1-14

# filtered = df[df['Ed_Q10'] == 13]
# filtered['Ed_Q10']
```

```python
## any point when Ed_Q10 and Ed_Q11 are the not the same and are not NaN? NO
different_values = df['Ed_Q10'].ne(df['Ed_Q11']) & df['Ed_Q11'].notna() & df['Ed_Q10'].notna()
result_df = df.loc[different_values, ['uid','Ed_Q10', 'Ed_Q11']]
assert(len(result_df) == 0 )

# The fact that result_df has no entries means we can just merge them!!!!
df['Merged_Q10_Q11'] = df['Ed_Q10'].fillna(df['Ed_Q11'])
df[['Merged_Q10_Q11']]
```
