### Which columns can be replaced by a zero?

* Categorical answered-questions for which zero is not one of the categories. By replacing them with a zero, we are implicitly adding an additional category for "No response"
  
* We need to make sure zero is not a feasible value in the column. otherwise it already means something and we need a new indicator. We also need to make sure it is categorical, if a column is numeric and we add a value of zero, then that doesnt make sense, add the average instead. 

```python
min_vals = df[columns_with_nan].min()
zero_valued_columns = []
non_zero_valued_columns = []

for index, value in min_vals.items():
    if float(value) < 1:
        zero_valued_columns.append(index)
        # print(f"Index: {index}, Value: {value}")
    else:
        non_zero_valued_columns.append(index)
```

# Merge Q9-11 and Q14-17

```python
df = pd.read_csv('education.csv') # education
df = prepare_df(df, "Ed")

RELEVANT_COLS = ['uid', 'Merged_Q8_Q9', 'Merged_Q10_Q11', 'Merged_Q14_Q15', 'Merged_Q16_Q17']
# Merge 10-11 and 16-17. Reasons for not attending/enrolling in school 

# first make them equivalent:
## the values of Q10 and Q11 match except for Q10=13 <-> Q11=14. So, Q10=13 should be transformed into Q10'=14 before merging 
## the values of Q16 and Q17 match except for Q16=13 <-> Q17=14. So, Q16=13 should be transformed into Q16'=14 before merging 

df.loc[df['Ed_Q10'] == 13, 'Ed_Q10'] = 14
df.loc[df['Ed_Q16'] == 13, 'Ed_Q16'] = 14


df['Merged_Q10_Q11'] = df['Ed_Q10'].fillna(df['Ed_Q11'])
df['Merged_Q16_Q17'] = df['Ed_Q16'].fillna(df['Ed_Q17'])
# df[['uid', 'Ed_Q08','Ed_Q09', 'Merged_Q10_Q11','Ed_Q14','Ed_Q15', 'Merged_Q16_Q17']].tail()
```

```python
# Claim: if Q9== Nan then Q08 == 2. so we can fill the NaN in attendance q9 with Q8 (enrolment)
different_values = df['Ed_Q09'].isna() & df['Ed_Q08'].eq(1) 
df.loc[different_values, ['uid', 'Ed_Q08', 'Ed_Q09']]
assert(len(df[different_values]) == 0)

# yup. So. 
df['Merged_Q8_Q9'] = df['Ed_Q09'].fillna(df['Ed_Q08'])
df['Merged_Q14_Q15'] = df['Ed_Q15'].fillna(df['Ed_Q14'])

# And 
df[RELEVANT_COLS].tail()
```

```python
# Now we are interested in how we will merge the case when someone did not attend school two years in a row for DIFFERENT REASONS: 
fltr = df['Merged_Q8_Q9'].eq(df['Merged_Q14_Q15']) & df['Merged_Q14_Q15'].eq(2.0) & df['Merged_Q10_Q11'].ne(df['Merged_Q16_Q17'])
# to conciliate reasons, we can actually only capture if the reason was financial or not!!!
df.loc[fltr, RELEVANT_COLS]
```

#### Conciliating reasons

the mapping of reason 1-13 -> how related that reason is to money on a scale of 1-3 is as follows: 

[Nan, 1, 2, 3, ..., 14] -> [0, 3,1,3,3,1,1,1,2,2,1,1,2,1,1]




```python
# Note since going to school really is kind of the opposite of being poor, we will assign  NaN reason values a mapping of 0

# Can we do that?
fltr = df['Merged_Q10_Q11'].eq(df['Merged_Q16_Q17']) & df['Merged_Q16_Q17'].isna() 
assert(len(df.loc[fltr, RELEVANT_COLS])==0)
# yes

# so here:
df['Merged_Q10_Q11'] = df['Merged_Q10_Q11'].fillna(0)
df['Merged_Q16_Q17'] = df['Merged_Q16_Q17'].fillna(0)

# Now apply the money transform (ONLY DO IT ONCE)
```

```python
# MERGE Merged_Q10_Q11 and Merged_Q16_Q17
# Now apply the money transform (ONLY DO IT ONCE)
ABSENCE_REASON_TO_MONEY_MAPPING = [0, 3,1,3,3,1,1,1,2,2,1,1,2,1,1]

df['Merged_Q10_Q11'] =  df['Merged_Q10_Q11'].apply(lambda indx : ABSENCE_REASON_TO_MONEY_MAPPING[int(indx)])
df['Merged_Q16_Q17'] =  df['Merged_Q16_Q17'].apply(lambda indx : ABSENCE_REASON_TO_MONEY_MAPPING[int(indx)])

# Now the merge between Merged_Q10_Q11 and Merged_Q10_Q11 becomes obvious: should be the maximum. 
df['Merged_Q10_Q11_Q16_Q17'] =  df.apply(lambda row: max(row['Merged_Q10_Q11'], row['Merged_Q16_Q17']), axis=1)

```

```python
# MERGE Merged_Q8_Q9 and Merged_Q14_Q15
# The meaning of this will be: 2 if in the last two years they have 'taken a sabbatical' been absent at least one. So basically is either one of them a 2?
two_selectors = df['Merged_Q8_Q9'].eq(2) | df['Merged_Q14_Q15'].eq(2)
df.loc[two_selectors, 'Merged_Q8_Q9_Q14_Q15'] = 2
df.loc[~two_selectors, 'Merged_Q8_Q9_Q14_Q15'] = 1

# df[['Merged_Q8_Q9', 'Merged_Q14_Q15', 'Merged_Q8_Q9_Q14_Q15']]
```

```python
RELEVANT_COLS = ['uid', 'Merged_Q8_Q9_Q14_Q15', 'Merged_Q10_Q11_Q16_Q17']

## CLAIM: Merged_Q10_Q11_Q16_Q17==0 <-> Merged_Q8_Q9_Q14_Q15=1
claim = df['Merged_Q8_Q9_Q14_Q15'].eq(1) & df['Merged_Q10_Q11_Q16_Q17'].ne(0)
assert(len(df.loc[claim, RELEVANT_COLS])==0)

# CLAIM: Merged_Q10_Q11_Q16_Q17 already captures all information. as it is ==0 when Merged_Q8_Q9_Q14_Q15==1.
df['MergedQ_8-11_14-17'] = df['Merged_Q10_Q11_Q16_Q17']

### DONE. FINAL TRANSFORM of Q8-11 and Q14-17. 
# It represents what happened in the last couple of years
# 0 if the kid attended both years to school
# 1 if the kid missed at least one year due to something (NOT) related to money
# 2 if the kid missed at least one year due to something (SOMEWHAT) related to money
# 3 if the kid missed at least one year due to something (DIRECTLY) related to money

RELEVANT_COLS = ['uid', 'MergedQ_8-11_14-17']  
df[RELEVANT_COLS].describe()
```
