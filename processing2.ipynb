# Feature engineering

```python
import pandas as pd
```

```python
# actualy, maybe get data before it is one hot encoded... too many columns
X = pd.read_csv('data/processed/X_processed.csv')
y = pd.read_csv('data/processed/y.csv')

```

```python
X.loc[X['time_away'] > 0, ['time_away']]
```

```python
X.columns
```

```python
X = X.drop(columns=['time_away', 'mother_age', 'father_age', 'highest_school_lvl', 'highest_school_lvl_grade'])
```

```python
X['male'] = X['sex'].eq(1)
X = X.drop(columns=['sex'])
```

```python
# They are only different in 208 cases 
X.loc[~X['now_not_enroll_reason'].eq(X['past_not_enroll_reason'])]

now_not_enroll_for_economic_reasons = X['now_not_enroll_reason'].isin([1,3,4,6,7,8,9,11])
past_not_enroll_for_economic_reasons = X['past_not_enroll_reason'].isin([1,3,4,6,7,8,9,11])

X['not_enrolled_for_economic_reasons'] = now_not_enroll_for_economic_reasons | past_not_enroll_for_economic_reasons
X = X.drop(columns=['now_not_enroll_reason', 'past_not_enroll_reason'])
```

```python
X
```

We will leave father_educationa and mother_aducationa as they are, without hot encoding them since they are already ordinal, ie highest numbered category corresponds to highest level of education.

Only categorical variables left are:

* marital_status
* family_role

```python
categorical_columns = ['marital_status', 'family_role']
X = pd.get_dummies(X, columns=categorical_columns)
# X = X.drop(columns=categorical_columns)
```
